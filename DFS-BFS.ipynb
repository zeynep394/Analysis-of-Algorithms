## Depth First Search
- Better solution to these problems.
- Used to make traversals on graphs.
- For the graphs we are investigating connectivity and acyclicity
- Based on decrease by one technique

Approach:
- Start from an irgendein node, mark it as visited
- On each iteration proceed to an unvisited adjacent node (If there are multiple unvisited adjacent nodes, which of them are chosen?)
- Continues until we reach a dead end (a node with no adjacent unvisited edges)
- When at a  dead end the algorithm backs up by one edge to the node it came from, and from there tries to find another unvisited adjacent node to continue traversing.
- Halts when it backs up until the starting node.
- If there are still unvisited nodes, DFS must be started from one of them. True for cases where we have graph forests.

Notes:
- Use a stack to store the visited nodes.
- push an unvisited node to the sack when it is first visited
- pop a node when it becomes a dead end.
- when a new unvisited node is reached it is added as a child to the node it was reached from
- back edge: the edge leading to a prev visited node

```python
%matplotlib inline
import matplotlib.pyplot as plt
import networkx as nx
from itertools import combinations
from random import random


def ER(n, p):
    V = set([v for v in range(n)])
    E = set()
    for combination in combinations(V, 2):
        a = random()
        if a < p:
            E.add(combination)

    g = nx.Graph()
    g.add_nodes_from(V)
    g.add_edges_from(E)

    return g


n = 5
p = 0.4
G = ER(n, p)
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos)
plt.title("Random Graph Generation Example")
plt.show()
```

### Analysis 

How efficient is DFS algorithm?
- It is dependent on the type of data structure used.
    - For adjancecy matrix representation time is in $\theta (|V|^{2})$
    - For adjancecy list representation time is in $\theta (|V| + |E|)$
